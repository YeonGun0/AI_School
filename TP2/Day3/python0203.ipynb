## 일자별 시세를 페이지별로 수집

```python
def get_day_list(item_code, page_no):
    """
    일자별 시세를 페이지별로 수집
    1) url 을 만든다.
    2) requests 를 통해 html 문서를 받음
    3) read_html을 통해 table 태그를 읽어 옴
    4) 결측행을 제거
    5) 데이터프레임 반환
    """
    
    # 일자별 시세를 출력하는 url
    url = f"https://finance.naver.com/item/sise_day.naver?code={item_code}&page={page_no}"
    
    # html 문서 받기
    res = requests.get(url, headers={"user-agent": "Mozilla/5.0"})
    
    # html 문서에서 table 태그 읽기
    table = pd.read_html(res.text)
    
    # 데이터 전처리
    df = table[0]
    df = df.dropna()
    df = df.reset_index(drop=True)
    
    return df
```

```python
# LG 에너지솔루션
item_code = 373220
item_name = "LG 에너지솔루션"
page_no = 1

get_day_list(item_code, page_no)
```

## 반복문으로 전체 일자 데이터 수집

```python
# 조은님 코드
item_code = "403550"
item_name = "쏘카"
page_no = 1

item_list = []
curr_day = None

while True:
    # 한 페이지의 일별 시세 수집
    df_item = get_day_list(item_code, page_no)
    # 해당 데이터의 마지막 날짜 가져옴
    last_day = df_item.iloc[-1]["날짜"]
    print(page_no, curr_day, last_day)
    
    # 해당 데이터의 마지막 날짜와 이전 데이터의 마지막 날짜를 비교
    # 맨 첫 데이터에는 이전 데이터 날짜가 없기 때문에 반복문 밖에서 초기화
    if last_day == curr_day:
        break
        
    curr_day = last_day
    item_list.append(df_item)
    page_no += 1
```

## 전체 과정을 하나의 함수로 만들기

```python
import time


def get_item_list(item_code, item_name):
    
    page_no = 1

    item_list = []
    pre_date = None

    while True:
        temp = get_day_list(item_code, page_no)

        # 날짜 추출
        cur_date = temp.iloc[-1][0]

        if pre_date == cur_date:
            break

        page_no += 1
        item_list.append(temp)
        pre_date = cur_date

    time.sleep(0.5)
    
    df_day = pd.concat(item_list)
    
    df_day["종목코드"] = item_code
    df_day["종목명"] = item_name
    
    # 파일 이름 위한 최신 날짜
    date = df_day.iloc[0]["날짜"]
    
    cols = ['종목코드', '종목명', '날짜', '종가', '시가', '고가', '저가', '거래량']
    df_day = df_day[cols]
    df_day = df_day.drop_duplicates()
          
    file_name = f"{item_name}_{item_code}_{date}.csv"
    df_day.to_csv(file_name, index=False)
    return file_name
```
