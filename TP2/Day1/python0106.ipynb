### 비대칭도 (왜도)
* 확률 분포 비대칭성을 나타내는 지표
* 왜도 값은 양수나 음수가 될 수 있고 정의되지 않을 수도 있음
* 음수일 경우 오른쪽에 더 많이 분포, 왼쪽에 긴 꼬리
* 양수일 경우 왼쪽에 더 많이 분포, 오른쪽에 긴 꼬리
* 평균과 중앙값이 같으면 왜도 = 0

```python
df.skew()
```

### 첨도
* 확률분포의 뾰족한 정도를 나타냄
* 얼마나 중심에 몰려 있는가를 측정
* 정규분포의 첨도 = 0
* 0보다 크면 더 뾰족한 모양
* 0보다 작으면 더 완만한 모양

```python
df.kurt()
```

### 상관분석
### 두 변수간의 연관된 정도를 나타낼 뿐 인과관계를 설명하는 것은 아니다

```python
corr = df.corr()
corr
```

```python
# corr의 shape에 1로 채운 배열
np.ones_like(corr)
```

```python
# np.triu(arr) : arr의 상삼각행렬
# np.tril(arr) : arr의 하삼각행렬
np.triu(np.ones_like(corr))
```

```python
mask = np.triu(np.ones_like(corr))
```

```python
# heatmap으로 상관계수 시각화
sns.heatmap(corr, cmap="coolwarm", annot=True)
```

```python
# 마스크 적용
# 중복값 제거
# mask : bool array or DataFrame, optional
#     If passed, data will not be shown in cells where ``mask`` is True.
#     Cells with missing values are automatically masked.
sns.heatmap(corr, cmap="coolwarm", annot=True, mask=mask)
```
