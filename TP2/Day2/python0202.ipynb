## url 반환 함수

```python
def get_url(item_code, page_no):
    url = f"https://finance.naver.com/item/news_news.nhn?code={item_code}&page={page_no}&sm=title_entity_id.basic&clusterId="
    return url
```

```python
# LG에너지솔루션 : 373220

print(get_url(373220, 1))
```

## 뉴스 한 페이지 수집하는 함수

```python
def get_one_page(item_code, page_num):
    url = get_url(item_code, page_num)
    
    table = pd.read_html(url)
    
    df_news = table[0]
    cols = table[0].columns
    
    temp_list = []
    for news in table[:-1]:
        news.columns = cols
        temp_list.append(news)
        
    # temp_list 합침
    df_news = pd.concat(temp_list)
    
    # 전처리
    df_news = df_news.dropna()
    df_news = df_news[~df_news["정보제공"].str.contains("연관기사")]
    df_news = df_news.drop_duplicates()
    
    # index 초기화
    df_news = df_news.reset_index(drop=True)
    
    return df_news
```

### page_no 까지 파라미터로 받아 함수 하나로 여러 페이지 데이터 수집

```python
from tqdm import trange
import time

item_code = df_krx.loc[df_krx["Name"] == "LG에너지솔루션", "Symbol"].values[0]
      
def get_many_pages(item_code, page_start, page_end):
    # news 수집 list
    temp_l = []
    
    # page_start부터 page_end까지 url 접근
    for page_num in trange(page_start, page_end + 1):
        url = get_url(item_code, page_num)
    
        table = pd.read_html(url)
        cols = table[0].columns
        
        # url 마다 news를 temp_l에 추가
        for news in table[:-1]:
            news.columns = cols
            temp_l.append(news)
        
        # 서버 부하 줄이기 위한 딜레이
        time.sleep(0.1)

    # temp_list df로 합침
    df_news1 = pd.concat(temp_l)
    
    # 전처리
    df_news1 = df_news1.dropna()
    df_news1 = df_news1[~df_news1["정보제공"].str.contains("연관기사")]
    df_news1 = df_news1.drop_duplicates()
    
    # index 초기화
    df_news1 = df_news1.reset_index(drop=True)
    
    return df_news1
```

```python
get_many_pages(item_code, 1, 31)
```
